## Classes

````python
class point:
    def __init__(self,coord_x=0,coord_y=0):
        self.x = int(coord_x)
        self.y = int(coord_y)
        
    def __rep__(self):
        return "X:{} Y:{}".format(self.x,self.y)
````

````python
pt = point(coord_y=4)
print(pt.y)
````
